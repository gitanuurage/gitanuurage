# Types of Operator

**Python language supports the following types of operators** −

    1.  Arithmetic Operators
    2.  Comparison (Relational) Operators
    3.  Assignment Operators
    4.  Logical Operators
    5.  Bitwise Operators
    6.  Membership Operators
    7.  Identity Operators

```python
a = 10
b = 20

print("Addition ", a+b)
print("Subtraction ", a-b)
print("Multiplication ", a*b)
print("Division ", a/b)

a = 107
b = 12

print("Modulus ", a%b)
print("Floor Division ", a//b)
print("Exponent ", a**b)
```

```python
a = 10
b = 20

print("a == b", a==b)
print("a != b", a!=b)
print("a > b ", a>b)
print("a >= b", a>=b)
print("a < b", a<b)
print("a <= b", a<=b)
```

```python
a = 10
b = 20

# a = a+b => a += b
a += b
print("a += b ", a)
a -= b
print("a -= b ", a)
a *= b
print("a *= b ", a)
a /= b
print("a /= b ", a)
a %= b
print("a %= b ", a)
a //= b
print("a //= b ", a)
a **= b
print("a **= b ", a)
```
