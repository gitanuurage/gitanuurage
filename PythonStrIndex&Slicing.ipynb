# Slicing String

image.png

image.png

# This slicing is called as positive slicing; as we are propagating\growing\pass on from left to right.

```python
var_str[0:5:1], var_str[6:12:1], var_str[13:19:1], var_str[20:24:1], var_str[25:31:1]
```

image.png

image.png

# Rules for positive indexing and positive slicing.

1. Default value of first element is zero; first element is smallest element.
2. Default value of second element is length of string; second element is largest element.
3. Default value of third element is always positive one (1); third element desides the flow, from left to right.
4. Rule for second element is (n - 1); where n is length of sub-string, after subtraction it return the "index value".
5. Rule for third element is (n - 1); where n is position, after subtraction it return the "step value"/skip value.

image.png

```python
print(var_str[0:5:1])
# output: Betty
# rule for 2nd element (n - 1); n = 5; 5 - 1 = 4 (index) 
# rule for 3rd element (n - 1); n = 1; 1 - 1 = 0 skip/jump/step

print(var_str[6:12:1])
# output: Botter
# rule for 2nd element (n - 1); n = 5; 5 - 1 = 4 (index) 
# rule for 3rd element (n - 1); n = 1; 1 - 1 = 0 skip/jump/step

print(var_str[13:19:1])
# output: bought
# rule for 2nd element (n - 1); n = 5; 5 - 1 = 4 (index) 
# rule for 3rd element (n - 1); n = 1; 1 - 1 = 0 skip/jump/step

print(var_str[20:24:1])
# output: some
# rule for 2nd element (n - 1); n = 5; 5 - 1 = 4 (index) 
# rule for 3rd element (n - 1); n = 1; 1 - 1 = 0 skip/jump/step

print(var_str[25:31:1])
# output: butter
# rule for 2nd element (n - 1); n = 5; 5 - 1 = 4 (index) 
# rule for 3rd element (n - 1); n = 1; 1 - 1 = 0 skip/jump/step
```

# Rules for negative indexing and negative slicing.

1.  Default value of first element is -ve 1; first element is largest element.
2.  Default value of second element is -ve length of string; second element is smallest element.
3.  Default value of third element is always negative one (-ve 1); third element desides the flow, from right to left.
4.  Rule for second element is (n + 1); where n is length of sub-string, after subtraction it return the "index value".
5.  Rule for third element is (n + 1); where n is position, after subtraction it return the "step value"/skip value.

```python
var_str[::-1]
```

image.png

```python
var_str[-1:-8:-1], var_str[-9:-13:-1], var_str[-14:-20:-1], var_str[-21:-27:-1], var_str[-28:-33:-1]
```

image.png

```python
print(var_str[-1:-8:-1])
# output: '.rettub'
# rule for 2nd element (n + 1); n = -8; -8 + 1 = -7 (index) 
# rule for 3rd element (n + 1); n = -1; -1 + 1 = 0 skip/jump/step

print(var_str[-9:-13:-1])
# output: 'emos'
# rule for 2nd element (n + 1); n = -13; -13 + 1 = -12 (index) 
# rule for 3rd element (n + 1); n =  -1; -1 + 1 = 0 skip/jump/step

print(var_str[-14:-20:-1])
# output: 'thguob'
# rule for 2nd element (n + 1); n = -20; -20 + 1 = -19 (index) 
# rule for 3rd element (n + 1); n =  1; -1 + 1 = 0 skip/jump/step

print(var_str[-21:-27:-1])
# output: 'rettoB'
# rule for 2nd element (n + 1); n = -27; -27 + 1 = -26 (index) 
# rule for 3rd element (n + 1); n =  1; -1 + 1 = 0 skip/jump/step

print(var_str[-28:-33:-1])
# output: 'ytteB'
# rule for 2nd element (n + 1); n = -33; -33 + 1 = -32 (index) 
# rule for 3rd element (n + 1); n =  1; -1 + 1 = 0 skip/jump/step
```
